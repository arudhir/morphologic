```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from IPython.display import display

from pathlib import Path
import pandas as pd
import numpy as np
import dask.dataframe as dd
import seaborn as sns
import matplotlib.pyplot as plt

data_dir = Path("/Volumes/T9/cellprofiler-output-egfr/cellprofiler-output/EGFR_EMPTY_control/")

def parse_result(fp, verbose=False):
    print(f'Parsing {fp}')
    df = dd.read_csv(fp, assume_missing=True)
    column_tags = df.columns.str.split('_').str[0].unique()
    
    print(f'Number of objects: {df.shape[0]}')
    print(f'Number of features: {df.shape[1]}')
    print(f'Column Tags: {column_tags}')
    print(f'Columns: {df.columns}')
    return df

def csv_to_ddf(fp):
    df = dd.read_csv(fp, assume_missing=True)
    return df

def filter_image_number(ddf, image_number=1.0):
    """
    Filters a Dask DataFrame to only include rows where 'ImageNumber' == image_number
    and converts it to a pandas DataFrame.

    Parameters:
    - ddf: Dask DataFrame to filter.
    - image_number: The ImageNumber value to filter by (default is 1.0).

    Returns:
    - pandas DataFrame filtered by the specified ImageNumber.
    """
    if 'ImageNumber' not in ddf.columns:
        print("ImageNumber column not found in the DataFrame")
        return
    filtered_ddf = ddf[ddf['ImageNumber'] == image_number]
    filtered_df = filtered_ddf.compute()
    return filtered_df

def dict_to_excel(dataframes_dict, filename):
    """
    Saves a dictionary of DataFrames (pandas or Dask) to a multi-sheet Excel file.
    
    Parameters:
    - dataframes_dict (dict): Dictionary where keys are sheet names and values are DataFrames (pandas or Dask).
    - filename (str): The path to the output Excel file.
    
    Returns:
    - None
    """
    import re
    import dask.dataframe as dd

    # Helper function to sanitize sheet names
    def sanitize_sheet_name(name):
        # Remove invalid characters
        invalid_chars = r'[:\\/?*[\]]'
        name = re.sub(invalid_chars, '_', str(name))
        # Truncate to 31 characters
        return name[:31]

    sheet_names = set()
    with pd.ExcelWriter(filename) as writer:
        for sheet_name, df in dataframes_dict.items():
            # Ensure the sheet_name is a string and sanitize it
            sheet_name = sanitize_sheet_name(sheet_name)
            # Ensure the sheet name is unique
            original_sheet_name = sheet_name
            counter = 1
            while sheet_name in sheet_names:
                sheet_name = f"{original_sheet_name}_{counter}"
                sheet_name = sheet_name[:31]  # Truncate if necessary
                counter += 1
            sheet_names.add(sheet_name)

            # Check if df is a Dask DataFrame
            if isinstance(df, dd.DataFrame):
                # Compute the Dask DataFrame to get a pandas DataFrame
                df_to_write = df.compute()
            elif isinstance(df, pd.DataFrame):
                # It's already a pandas DataFrame
                df_to_write = df
            elif df is None:
                print(f"Skipping sheet '{sheet_name}' because it is None")
                continue
            else:
                # If df is neither pandas nor Dask DataFrame, raise an error
                raise ValueError(f"The object for sheet '{sheet_name}' is not a pandas or Dask DataFrame.")

            # Write the pandas DataFrame to the Excel sheet
            df_to_write.to_excel(writer, sheet_name=sheet_name, index=False)
            
# Load the CSVs into Dask DataFrames
cp_files = {
    # 'Image': 'Recursion_6Ch_Image.csv',
    'Cells': csv_to_ddf(data_dir / 'Recursion_6Ch_Cells.csv'),
    # 'Experiment': csv_to_ddf(data_dir / 'Recursion_6Ch_Experiment.csv'),
    'Cytoplasm': csv_to_ddf(data_dir / 'Recursion_6Ch_Cytoplasm.csv'),
    'GolgiSpots': csv_to_ddf(data_dir / 'Recursion_6Ch_GolgiSpots.csv'),
    'MitoObjects': csv_to_ddf(data_dir / 'Recursion_6Ch_MitoObjects.csv'),
    'MitoChildObjects': csv_to_ddf(data_dir / 'Recursion_6Ch_MitoChildObjects.csv'),
    'Nuclei': csv_to_ddf(data_dir / 'Recursion_6Ch_Nuclei.csv'),
    'Nucleoli': csv_to_ddf(data_dir / 'Recursion_6Ch_Nucleoli.csv'),
    'NucleoliChildObjects': csv_to_ddf(data_dir / 'Recursion_6Ch_NucleoliChildObjects.csv'),
}


cp_files['Cells'].head()
```

# Get image-level means

```python
import numpy as np
def mad_scaled_median(series):
    median = np.median(series)
    mad = np.median(np.abs(series - median))
    return median / mad if mad != 0 else median  # scale by MAD, avoid division by zero

def image_aggregate(dask_df, image_col='ImageNumber', custom_func=mad_scaled_median, shuffle_method='tasks'):
    # select numeric columns only
    numeric_df = dask_df.select_dtypes(include=[np.number])
    
    # apply custom aggregation function to each column
    agg_df = numeric_df.groupby(image_col).apply(lambda df: df.apply(custom_func), meta=numeric_df._meta)
    
    return agg_df.compute()
```

```python
image_aggregates = {}
for experiment in ['EGFR_EMPTY_control', 'EGFR_KO']:
    data_dir = Path(f"/Volumes/T9/cellprofiler-output-egfr/cellprofiler-output/{experiment}/")
    cells_df = csv_to_ddf(data_dir / 'Recursion_6Ch_Cells.csv')
    image_aggregates[experiment] = image_aggregate(cells_df, custom_func=mad_scaled_median, shuffle_method='tasks')
```

# Do a RankSum test for each feature

Get which features are significantly different between the two groups

```python
from scipy.stats import ranksums
def compare_all_features_ranksum(df1, df2):
    # get common columns that are numeric
    common_columns = [col for col in df1.select_dtypes(include=[np.number]).columns if col in df2.columns]
    
    # list to store results
    results = []
    
    for column in common_columns:
        # dropna for the two columns being compared
        data1 = df1[column].dropna()
        data2 = df2[column].dropna()
        
        # perform rank-sum test and store p-value
        _, pval = ranksums(data1, data2)
        results.append({'feature_name': column, 'significance': pval})
    
    # create DataFrame from results
    return pd.DataFrame(results)

# usage example
significance_df = compare_all_features_ranksum(image_aggregates['EGFR_EMPTY_control'], image_aggregates['EGFR_KO'])

significance_df.dropna(subset=['significance']).sort_values(by='significance').head(40)
```
