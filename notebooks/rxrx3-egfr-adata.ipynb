```python
import pandas as pd
import numpy as np
from pycytominer import normalize, feature_select, aggregate
from pycytominer.cyto_utils import output

def setup_analysis_config():
    config = {
        'input_path': '/Volumes/T9/cellprofiler-output-egfr-first/cellprofiler-output/EGFR_KO/Recursion_6Ch_*.csv',  # Update with your path
        'output_path': '/Volumes/T9/cellprofiler-output-egfr-first/pycytominer',
        'metadata_cols': [
            'Metadata_Plate',
            'Metadata_Well',
            'Metadata_Site',
            'ImageNumber'
        ],
        'exclude_features': [
            "ObjectNumber",
            "Location_Center_X",
            "Location_Center_Y",
            "Mouse_X",
            "Mouse_Y"
        ],
        'group_cols': ['Metadata_Plate', 'Metadata_Well'],
        # Subsampling parameters
        'cells_per_well': 100,  # Number of cells to keep per well
        'min_cells_per_well': 50,  # Minimum cells required for a well to be included
        'quality_features': [
            'Cells_AreaShape_Area',
            'Nuclei_AreaShape_Area',
            'Cells_AreaShape_Perimeter',
            'Nuclei_Intensity_MeanIntensity_DNA'
        ]
    }
    return config

def filter_cells_by_quality(df, quality_features):
    """
    Remove outlier cells based on key quality metrics
    """
    filtered_df = df.copy()
    
    for feature in quality_features:
        if feature in df.columns:
            # Calculate quality metrics
            Q1 = df[feature].quantile(0.1)
            Q3 = df[feature].quantile(0.9)
            IQR = Q3 - Q1
            
            # Filter cells within reasonable bounds (extended IQR range)
            mask = (df[feature] >= (Q1 - 2 * IQR)) & (df[feature] <= (Q3 + 2 * IQR))
            filtered_df = filtered_df[mask]
    
    return filtered_df

def subsample_data(df, config):
    """
    Intelligently subsample the dataset while maintaining representation
    """
    print(f"Original dataset size: {len(df)}")
    
    # First filter by quality
    filtered_df = filter_cells_by_quality(df, config['quality_features'])
    print(f"Size after quality filtering: {len(filtered_df)}")
    
    # Group by well
    grouped = filtered_df.groupby([config['group_cols'][0], config['group_cols'][1]])
    
    sampled_dfs = []
    for name, group in grouped:
        # Only include wells with sufficient cells
        if len(group) >= config['min_cells_per_well']:
            # Stratified sampling based on nuclear area quartiles to maintain morphology distribution
            if 'Nuclei_AreaShape_Area' in group.columns:
                group['Area_Quartile'] = pd.qcut(group['Nuclei_AreaShape_Area'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
                sampled = group.groupby('Area_Quartile', group_keys=False).apply(
                    lambda x: x.sample(min(len(x), config['cells_per_well']//4))
                )
                sampled_dfs.append(sampled)
            else:
                # Simple random sampling if area measurement not available
                sampled = group.sample(min(len(group), config['cells_per_well']))
                sampled_dfs.append(sampled)
    
    subsampled_df = pd.concat(sampled_dfs, ignore_index=True)
    print(f"Final subsampled size: {len(subsampled_df)}")
    
    # Calculate and print sampling statistics
    well_counts = subsampled_df.groupby(config['group_cols']).size()
    print(f"\nSampling Statistics:")
    print(f"Number of wells: {len(well_counts)}")
    print(f"Average cells per well: {well_counts.mean():.1f}")
    print(f"Min cells per well: {well_counts.min()}")
    print(f"Max cells per well: {well_counts.max()}")
    
    return subsampled_df

def load_and_merge_data(config):
    """
    Load CellProfiler output CSV files and merge if multiple files exist
    """
    import glob
    
    csv_files = glob.glob(config['input_path'])
    dfs = []
    
    for file in csv_files:
        df = pd.read_csv(file)
        dfs.append(df)
    
    if len(dfs) > 1:
        merged_df = pd.concat(dfs, ignore_index=True)
    else:
        merged_df = dfs[0]
    
    return merged_df

# ... [rest of the previous functions remain the same] ...

def main():
    # Setup configuration
    config = setup_analysis_config()
    
    # Load data
    print("Loading data...")
    df = load_and_merge_data(config)
    
    # Subsample data
    print("\nSubsampling data...")
    subsampled_df = subsample_data(df, config)
    
    # Preprocess data
    print("\nPreprocessing data...")
    processed_df = preprocess_data(subsampled_df, config)
    
    # Aggregate profiles
    print("Aggregating profiles...")
    aggregated_df = aggregate_profiles(processed_df, config)
    
    # Analyze specific features
    print("Analyzing features...")
    analysis_results = analyze_specific_features(aggregated_df)
    
    # Save results
    print("Saving results...")
    output_path = config['output_path']
    
    import os
    os.makedirs(output_path, exist_ok=True)
    
    processed_df.to_csv(f"{output_path}processed_profiles.csv", index=False)
    aggregated_df.to_csv(f"{output_path}aggregated_profiles.csv", index=False)
    
    for key, result in analysis_results.items():
        if isinstance(result, dict):
            for sub_key, sub_result in result.items():
                sub_result.to_csv(f"{output_path}{key}_{sub_key}_analysis.csv")
        else:
            result.to_csv(f"{output_path}{key}_analysis.csv")
    
    print("Analysis complete!")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from pathlib import Path

def read_all_embeddings(base_path):
    """
    Reads all embeddings.parquet files from a nested folder structure
    and combines them into a single DataFrame with additional metadata.
    """
    base_path = Path(base_path)
    all_files = base_path.glob("embeddings/*/Plate*/embeddings.parquet")
    data = []

    for file in all_files:
        parts = file.parts
        compound = parts[-3]
        plate = parts[-2]
        df = pd.read_parquet(file)
        # df['compound'] = compound
        # df['plate'] = plate
        data.append(df)

    return pd.concat(data, ignore_index=True)

embeddings = read_all_embeddings('../data/embeddings/')
embeddings.set_index('well_id', inplace=True)
embeddings.head()
```
